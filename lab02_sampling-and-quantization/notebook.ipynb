**Step 2: Load and Display an Image**
1. Import necessary libraries:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

2. Upload an Image to the Collab Session

```python
from google.colab import files
uploaded = files.upload()

# NOTE: image used is https://raw.githubusercontent.com/SiegfredLorelle/digital-image-processing/refs/heads/main/lab02_sampling-and-quantization/data/us.jpg
```

3. Load the uploaded image and display it:

```python
image = cv2.imread("us.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

**Step 3: Sampling**

Task 1: Reduce Image Resolution
1. Downsample the image by reducing its size:

```python
downsampled = cv2.resize(image_rgb,(image_rgb.shape[1]//2, image_rgb.shape[0]))
plt.imshow(downsampled)
plt.title("Downsampled Image")
plt.axis("off")
plt.show()
```

2. Experiment with different resolution scales (e.g., 1/4, 1/8) and observe the effects.

```python
downsampled_8 = cv2.resize(image_rgb,(image_rgb.shape[1]//8, image_rgb.shape[0]))
plt.imshow(downsampled_8)
plt.title("Downsampled Image")
plt.axis("off")
plt.show()
```

Task 2: Restore Original Resolution
1. Resize the downsampled image back to its original dimensions:

```python
restored = cv2.resize(downsampled, (image_rgb.shape[1], image_rgb.shape[0]))
plt.imshow(restored)
plt.title("Restored Image")
plt.axis("off")
plt.show()
```

2. Compare the restored image to the original image and document the differences.

The restored image is blurry due to the restoration from being compressed and then stretched out without enough pixels.

**Step 4: Quantization**

Task 1: Reduce Color Depth

1. Perform color quantization by reducing the number of bits per channel:

```python
levels = 4 # Experiment with 2, 4, 8, etc.
quantized = np.floor(image_rgb / (256 // levels)) * (256 // levels)
quantized = quantized.astype(np.uint8)
plt.imshow(quantized)
plt.title(f"Quantized Image with {levels} Levels")
plt.axis("off")
plt.show()
```

2. Try different levels of quantization and observe the changes in image quality

```python
levels = 8 # Experiment with 2, 4, 8, etc.
quantized = np.floor(image_rgb / (256 // levels)) * (256 // levels)
quantized = quantized.astype(np.uint8)
plt.imshow(quantized)
plt.title(f"Quantized Image with {levels} Levels")
plt.axis("off")
plt.show()
```

```python
levels = 16 # Experiment with 2, 4, 8, etc.
quantized = np.floor(image_rgb / (256 // levels)) * (256 // levels)
quantized = quantized.astype(np.uint8)
plt.imshow(quantized)
plt.title(f"Quantized Image with {levels} Levels")
plt.axis("off")
plt.show()
```

Task 2: Analyze the Trade-offs

1. Discuss the trade-offs between color depth and image quality in your observations

By increasing the color depth, the image colors improves getting closer to the original color quality of the image.

**Step 5: Combined Effects**
1. Combine sampling and quantization to the image


*   Downsample the image first.

```python
combined = np.floor(downsampled / (256 // levels)) * (256 // levels)
combined = combined.astype(np.uint8)
plt.imshow(combined)
plt.title("Downsampled and Quantized Image")
plt.axis("off")
plt.show()
```
